 **Uniforme** $X \sim U(0,1)$

Funcion de densidad

\begin{equation}
f(x)=
\begin{cases}
\frac{1}{b-a} & \text{si } x \in (a, b) \\
0 & \text{si } x \notin (a, b)
\end{cases}
\end{equation}

image.png

Funcion acumulativa

\begin{equation}
F(x)=
\begin{cases}
0 & \text{si } x \leq a  \\
\frac{x - a}{b - a} & \text{si } x \in (a, b) \\
1 & \text{si } x \geq b
\end{cases}
\end{equation}


Image13.gif

Media

\begin{equation}
\mu = \frac{a+b}{2}, \quad a<x<b
\end{equation}

Varianza

\begin{equation}
\sigma^2 = \frac{(b-a)^2}{12}, \quad a<x<b
\end{equation}

Desviacion estandar

\begin{equation}
\sigma = \frac{(b-a)}{\sqrt{12}}, \quad a<x<b
\end{equation}

**Pseudocodigo**



1.   Generar $U \sim U(0,1)$
2.   Definir $x=a+(b-a)U$
3.   Retornar las x`s


```python
#UNIFORME
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función de la transformada inversa para la distribucion Uniforme
def uniforme(n):
  a,b=1,3 #Parametros
  x = [] #Lista vacia
  for i in range(n):
    R = random() #Numeros Aleatorios
    x.append(a+(b-a)*R)
  return x

#Muestras
A = uniforme(100000)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.5, edgecolor="black")
plt.title("Distribución Uniforme")


#Media y varianza
print("Media:", np.mean(A))
print("Varianza:", np.var(A))
print("Desviación estándar:", np.std(A))
```

 **Triangular** $X \sim U(0,1)$

Funcion de densidad
\begin{equation}
f(x) =
\begin{cases}
\displaystyle \frac{2(x - a)}{(b - a)(c - a)} & \text{para } a \leq x \leq c \\[10pt]
\displaystyle \frac{2(b - x)}{(b - a)(b - c)} & \text{para } c < x \leq b
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x) =
\begin{cases}
\displaystyle \frac{(x - a)^2}{(b - a)(c - a)} & \text{para } a \leq x \leq c \\[10pt]
\displaystyle 1 - \frac{(b - x)^2}{(b - a)(b - c)} & \text{para } c < x \leq b
\end{cases}
\end{equation}

image.png

Media

\begin{equation}
\mu = \frac{a+b+c}{3}, \quad a<x<b
\end{equation}

Varianza

\begin{equation}
\sigma^2 = \frac{a^2+b^2+c^2-ab-ac-bc}{18}, \quad a<x<b
\end{equation}

Desviacion estandar

\begin{equation}
\sigma = \sqrt{\frac{a^2+b^2+c^2-ab-ac-bc}{18}}, \quad a<x<b
\end{equation}

**Pseudocodigo**



1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = \begin{cases}
a + \sqrt{R \cdot (b - a)(c - a)} & \text{si } R < \frac{b-a}{c-a} \\
c - \sqrt{(1 - R) \cdot (c - b)(c - a)} & \text{si } R \geq \frac{b-a}{c-a}
\end{cases} $$
3.   Retornar las x`s


```python
#TRIANGULAR
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función de la transformada inversa para la distribucion triangular
def triangular(n):
  a,b,c=1,3,5 #Parametros
  x = [] #Lista vacia
  for i in range(n):
    R = random() #Numeros Aleatorios
    if R<(b-a)/(c-a):
      x.append(a + np.sqrt(R * (b - a) * (c - a)))
    elif R>=(b-a)/(c-a):
      x.append(c - np.sqrt((1 - R) * (c - b) * (c - a)))
  return x

#Muestras
A = triangular(10000)

#Histograma
plt.hist(A, bins=100, density=True, color='green', edgecolor='black', alpha=0.5)
plt.title("Distribución Triangular")
plt.show()

#Media y varianza
print("Media:", np.mean(A))
print("Varianza:", np.var(A))
print("Desviación estándar:", np.std(A))
```

\begin{equation}

**Pseudocidigo**

1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = -\frac{In(U)}{\lambda} $$

```python
#EXPONENCIAL
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from random import random

def exponencial(lam, n):
  x = []
  for i in range(n):
    R = random()
    x.append(-np.log(R)/lam)
  return x

exp = exponencial(1.5,100000)

plt.hist(exp, bins=100, density=True, alpha=0.5, edgecolor='black')
plt.show()

#Resultados teóricos y empíricos
lam = 1.5
media_teorica = 1 / lam
varianza_teorica = 1 / (lam** 2)
std_teorica = np.sqrt(varianza_teorica)

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {std_teorica:.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media empirica: {np.mean(exp):.4f}")
print(f"varianza empirica: {np.var(exp):.4f}")
print(f"Desviacion estandar empirica:{np.std(exp):.4f}" )
```

 **m-Erlang** $X \sim Erlang(n,\lambda)$

- $n \in \mathbb{N}$
-$\lambda > 0$

Funcion de densidad

\begin{equation}
 f(n, \lambda) = \frac{\lambda (\lambda x)^{n-1} e^{-\lambda x}}{(n-1)!}, \quad x \geq 0
\end{equation}

image.png

Funcion de distribucion

\begin{equation}
F(x) = 1 - \sum_{k=0}^{n-1} \frac{(\lambda x)^k e^{-\lambda x}}{k!}, \quad x \geq 0
\end{equation}

image.png

Media

\begin{equation}
\mu =  \frac{n}{\lambda}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{n}{\lambda^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{\sqrt{n}}{\lambda}
\end{equation}

**Pseudocodigo**

1.   Generar $U_1, U_2, ..., U_m \quad$ con $\quad U_i \sim U(0,1)$
2.   Definir $$ x = -\frac{1}{\lambda} \ln\left( \prod_{i=1}^m U_i \right) $$
3.   Retornar las x`s



```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

def erlang(m, lam, n):
    x = []
    for _ in range(n):
        u = [random() for _ in range(m)]  #Generar m números aleatorios uniformes
        P = np.prod(u)                    #Calcular el producto de los U_i
        x.append(-np.log(P) / lam)        #Aplicar la transformada inversa
    return x

#Parámetros
m = 3       #Número de etapas (eventos)
lam = 2     #Tasa (parámetro de escala)
n = 10000   #Número de muestras

#Generar muestras
A = erlang(m, lam, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.5, edgecolor="black", color = "orange")
plt.title(f"Distribución de m-Erlang")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()

#Media, varianza y desviación estándar teóricas
media_teorica = m / lam
varianza_teorica = m / (lam ** 2)
desviacion_teorica = np.sqrt(varianza_teorica)

#Resultados
print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}")

print("\nResultados numericos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

 **Gamma** $X \sim Gamma(\alpha, \beta)$

 Función de densidad
\begin{equation}
f(x; \alpha, \beta) =
\begin{cases}
\frac{\beta^\alpha x^{\alpha - 1} e^{-\beta x}}{\Gamma(\alpha)} & \text{si } x > 0 \\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x; \alpha, \beta) = \frac{\gamma(\alpha, \beta x)}{\Gamma(\alpha)}
\end{equation}

image.png

Media

\begin{equation}
\mu =  \frac{\alpha}{\beta}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{\alpha}{\beta^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{\sqrt{\alpha}}{\beta}
\end{equation}

**Pseudocodigo**

Dependiendo del valor de $\alpha$:

#### **Caso $\alpha = 1$ (Distribución Exponencial)**  
1. Generar $U \sim U(0,1)$  
2. Calcular $x = -\frac{\ln(U)}{\beta}$  
3. Retornar $x$

#### **Caso $0 < \alpha < 1$ (Algoritmo GS de Ahrens-Dieter)**  
1. Calcular $b = \frac{e + \alpha}{e}$  
2. Generar $U_1, U_2 \sim U(0,1)$  
3. Si $bU_1 \leq 1$:  
   - Calcular $Y = (bU_1)^{1/\alpha}$  
   - Si $U_2 \leq e^{-Y}$, retornar $x = Y/\beta$  
4. Si no:  
   - Calcular $Y = -\ln\left(\frac{b - bU_1}{\alpha}\right)$  
   - Si $U_2 \leq Y^{\alpha-1}$, retornar $x = Y/\beta$  
5. Repetir desde 2 si no se acepta

#### **Caso $\alpha > 1$ (Algoritmo GB de Cheng)**  
1. Calcular constantes:  
   - $a = \frac{1}{\sqrt{2\alpha - 1}}$  
   - $b = \alpha - \ln(4)$  
   - $q = \alpha + \frac{1}{a}$  
   - $\theta = 4.5$  
   - $d = 1 + \ln(\theta)$  
2. Generar $U_1, U_2 \sim U(0,1)$  
3. Calcular:  
   - $V = a \ln\left(\frac{U_1}{1-U_1}\right)$  
   - $Y = \alpha e^V$  
   - $Z = U_1^2 U_2$  
   - $W = b + qV - Y$  
4. Si $W + d - \theta Z \geq 0$, retornar $x = Y/\beta$  
5. Si $W \geq \ln(Z)$, retornar $x = Y/\beta$  
6. Repetir desde 2 si no se acepta


```python
#GAMMA
#Librerias
import numpy as np
from math import gamma, exp, log, sqrt
from scipy.stats import gamma as gamma_dist
import matplotlib.pyplot as plt

def gamma_ahrens_dieter(alpha, n_samples):
    """
    Implementación del algoritmo GS de Ahrens-Dieter para generar variables Gamma(α,1)
    con 0 < α < 1.

    Parámetros:
        alpha: parámetro de forma (0 < alpha < 1)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,1)
    """
    samples = []
    b = (np.e + alpha)/np.e  #Constante b

    while len(samples) < n_samples:
        U1 = np.random.uniform()
        P = b * U1

        if P <= 1:
            #Paso 2 del algoritmo
            Y = P ** (1/alpha)
            U2 = np.random.uniform()
            if U2 <= exp(-Y):
                samples.append(Y)
        else:
            #Paso 3 del algoritmo
            Y = -log((b - P)/alpha)
            U2 = np.random.uniform()
            if U2 <= Y ** (alpha - 1):
                samples.append(Y)

    return np.array(samples)

#Implementación del Algoritmo GB de Cheng para Gamma(α,1) con α > 1

def gamma_cheng(alpha, n_samples):
    """
    Implementación del algoritmo GB de Cheng para generar variables Gamma(α,1)
    con α > 1.

    Parámetros:
        alpha: parámetro de forma (alpha > 1)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,1)
    """
    samples = []
    a = 1/sqrt(2*alpha - 1)
    b = alpha - log(4)
    q = alpha + 1/a
    theta = 4.5
    d = 1 + log(theta)

    while len(samples) < n_samples:
        U1, U2 = np.random.uniform(size=2)
        V = a * log(U1/(1 - U1))
        Y = alpha * exp(V)
        Z = U1**2 * U2
        W = b + q*V - Y

        #Paso 3: pretest
        if W + d - theta*Z >= 0:
            samples.append(Y)
            continue

        #Paso 4: test completo
        if W >= log(Z):
            samples.append(Y)

    return np.array(samples)

#Función general para generar Gamma(α,β)

def generar_gamma(alpha, beta=1, n_samples=1):
    """
    Generador general de variables Gamma(α,β) que selecciona el algoritmo apropiado.

    Parámetros:
        alpha: parámetro de forma (alpha > 0)
        beta: parámetro de escala (beta > 0)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,β)
    """
    if alpha <= 0 or beta <= 0:
        raise ValueError("alpha y beta deben ser positivos")

    if alpha == 1:
        #Caso exponencial
        samples = -np.log(np.random.uniform(size=n_samples)) * beta
    elif alpha < 1:
        #Algoritmo de Ahrens-Dieter para 0 < α < 1
        samples = gamma_ahrens_dieter(alpha, n_samples) * beta
    else:
        #Algoritmo de Cheng para α > 1
        samples = gamma_cheng(alpha, n_samples) * beta

    return samples

#Ejemplo
#Parámetros
alpha = 0.7  # Prueba con valores <1 y >1
beta = 2.0
n_samples = 100000

#Muestras
samples = generar_gamma(alpha, beta, n_samples)

plt.figure(figsize=(10, 6))
counts, bins, _ = plt.hist(samples, bins=100, density=True,
                          alpha=0.7, edgecolor="black")

# PDF teórica para comparación
x_teorico = np.linspace(0, np.max(samples), 1000)
pdf_teorica = gamma_dist.pdf(x_teorico, a=alpha, scale=beta)

plt.title(f'Distribución Gamma')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()

# Estadísticas
media_teorica = alpha * beta
varianza_teorica = alpha * beta**2

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {sqrt(varianza_teorica):.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media muestral: {np.mean(samples):.4f}")
print(f"Varianza muestral: {np.var(samples):.4f}")
print(f"Desviación estándar muestral: {np.std(samples):.4f}")
```

**Weibull** $X \sim We(k,\lambda)$

Función de densidad
\begin{equation}
f(x)=
\begin{cases}
\lambda \alpha (\lambda x)^{\alpha-1}e^{-(\lambda x)\alpha}  & \text{si } x > 0\\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x)=
\begin{cases}
1 - e^{-(\lambda x)\alpha}  & \text{si } x > 0\\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Media

\begin{equation}
\mu =  \frac{1}{\lambda} \, \Gamma\left(1 + \frac{1}{\alpha}\right)
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{1}{\lambda^2} \left[ \Gamma \left( 1 + \frac{2}{\alpha} \right) - \Gamma^2 \left( 1 + \frac{1}{\alpha} \right) \right]
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sqrt{\frac{1}{\lambda^2} \left[ \Gamma \left( 1 + \frac{2}{\alpha} \right) - \Gamma^2 \left( 1 + \frac{1}{\alpha} \right) \right]}
\end{equation}

**Pseudocidigo**

1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = \beta -In(U)^{1/\alpha} $$
3.   Retornar las x`s



```python
#WEIBULL
import numpy as np
import matplotlib.pyplot as plt
from math import gamma

def generar_weibull(alpha, beta, n):
    U = np.random.uniform(0, 1, n)        #Genera n números aleatorios U(0,1)
    x = beta * (-np.log(U)) ** (1/alpha)  #Transformada inversa Weibull
    return x

#Parámetros de la distribución Weibull
alpha = 2.0  #Parámetro de forma (k)
beta = 1.5    #Parámetro de escala (λ)
n = 100000    #Número de muestras

#Generar datos
datos = generar_weibull(alpha, beta, n)

#Grafica
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=100, density=True, color="green", alpha=0.5, edgecolor="black", label='Datos simulados')

#PDF teórica para comparación
x_teorico = np.linspace(0, np.max(datos), 1000)
pdf_teorica = (alpha/beta) * (x_teorico/beta)**(alpha-1) * np.exp(-(x_teorico/beta)**alpha)

plt.title(f"Distribución Weibull")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")
plt.show()

#Cálculos teóricos y empíricos
media_teorica = beta * gamma(1 + 1/alpha)
varianza_teorica = beta**2 * (gamma(1 + 2/alpha) - (gamma(1 + 1/alpha))**2)

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {np.sqrt(varianza_teorica):.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media muestral: {np.mean(datos):.4f}")
print(f"Varianza muestral: {np.var(datos):.4f}")
print(f"Desviación estándar muestral: {np.std(datos):.4f}")
```

 **Normal** $X \sim N(\mu, \sigma)$

Funcion de densidad
\begin{equation}
f(x) = \frac{1}{\sqrt{2\pi}\sigma} \, e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x) = \frac{1}{2} \left[ 1 + erf\left( \frac{x-\mu}{\sqrt{2}\sigma}\right) \right]
\end{equation}

image.png

Media

\begin{equation}
\mu =  \mu
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \sigma^2
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sigma
\end{equation}

**Pseudocodigo**

1.   Generar $U_1 \sim U(0,1)$
2.   Definir $$ x = \frac{U^{0.135} - (1-U)^{0.135}}{0.1975} $$
3.   Retornar las x`s

```python
#NORMAL
#Librerias
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math

def distr_normal(n):
  lista = []
  for i in range(n):
    R=rd.random()
    x=float((R**0.135-(1-R)**0.135)/0.1975)
    lista.append(x)
  return lista

m = distr_normal(100000)

plt.hist(m, bins = 50, density=True, color="red", alpha=0.5, edgecolor="black")
plt.show()

media = np.mean(m)
print("La media es: ", media)

varianza = np.var(m)
print("La varianza es: ", varianza)
```

por Box-Muller

```python
#Librerias
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math

#Método de Box-Muller
def box_muller(n):
    lista = []
    for i in range(n):
        U1 = rd.random()
        U2 = rd.random()
        R = math.sqrt(-2 * math.log(U1))
        t = 2 * math.pi * U2
        x = R * math.cos(t)
        y = R * math.sin(t)
        lista.append(x)
        lista.append(y)
    return lista

#Muestras
m = box_muller(100000)  #Genera 200000 números

#Grafica
plt.hist(m, bins = 50, density=True, color="red", alpha=0.5, edgecolor="black")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.title("Distribución Normal")
plt.show()

#Estadísticas
media = np.mean(m)
desviacion = np.std(m)
varianza = np.var(m)

print("\n--- Resultados empíricos ---")
print("Media obtenida: ", media)
print("Varianza obtenida: ", varianza)
print("Desviación estándar obtenida: ", desviacion)

print("\n--- Resultados teóricos ---")
print("Media teórica: 0")
print("Varianza teórica: 1")
print("Desviación estándar teórica: 1")
```

**Log-normal**

$X \sim \mathcal{LN}(\mu,\sigma^2) \qquad \sigma>0 \qquad e^\mu >0 \qquad x \in (-∞, ∞)$

f(x) =
\begin{cases}
\frac{1}{\sigma x \sqrt{2\pi}} e^{-\frac{(ln \quad x-\mu)^2}{2\sigma^2}} & \text{para } x > 0 \\
0 & \text{otro caso}
\end{cases}
\end{equation}

Media

$$\mu=e^{\mu + \sigma^2 /2}$$

Varianza
$$\sigma^2=e^{\mu + \sigma^2 /2}(e^{\sigma^2}-1)$$

Desviacion estandar
$$\sigma^2=\sqrt{e^{\mu + \sigma^2 /2}(e^{\sigma^2}-1)}$$

**Pseudocodigo**

con $E[X]=\mu'= e^{\mu + \frac{\sigma^2}{2}} \qquad V[X]=\sigma'^2=(e^{\sigma^2} - 1) e^{2\mu + \sigma^2}$

Entonces

$E[Y]=\mu=ln( \frac{\mu´^2}{\sqrt{\mu´^2+\sigma´^2}})$

$V[Y]\sigma^2=ln(1+\frac{\sigma´^2}{\mu´^2})$

1. Generar $Y \sim N(\mu, \sigma^2)$
2. return $X=e^Y$

```python
#LOG-NORMAL
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def generar_lognormal(n, mu_ln, sigma_ln):

    #Convertir parámetros Log-Normales a parámetros Normales subyacentes
    sigma2_y = np.log(1 + (sigma_ln**2) / (mu_ln**2))
    mu_y = np.log(mu_ln) - sigma2_y / 2

    #Generar Y ~ N(mu_y, sigma2_y)
    Y = np.random.normal(mu_y, np.sqrt(sigma2_y), n)

    #Transformar a Log-Normal: X = e^Y
    X = np.exp(Y)

    return X

#Parámetros
mu_ln = 2.0     #Media deseada de la distribución Log-Normal
sigma_ln = 1.5  #Desviación estándar deseada de la distribución Log-Normal
n = 10000       #Número de muestras

#Generar muestras
muestras = generar_lognormal(n, mu_ln, sigma_ln)

#Calcular estadísticos empíricos
media_emp = np.mean(muestras)
varianza_emp = np.var(muestras)
std_emp = np.std(muestras)

#Histograma
plt.figure(figsize=(10, 5))
plt.hist(muestras, bins=100, density=True, color="orange", alpha=0.5, edgecolor="black")
x = np.linspace(0, 20, 1000)
plt.title(f"Distribución Log-Normal")
plt.legend()
plt.show()

# Resultados
print("\n--- Resultados empíricos ---")
print(f"Media obtenida: {media_emp:.4f}")
print(f"Desviación estándar obtenida: {std_emp:.4f}")
print(f"Varianza obtenida: {varianza_emp:.4f}")


print("\n--- Resultados teóricos ---")
print("Media teórica: 2")
print("Desviación estándar teórica: 1.5")
print("Varianza teórica: 2.25")
```

Por Box-Muller

```python
# Librerías
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math

#Método de Box-Muller para generar normales
def box_muller(n):
    lista = []
    for _ in range(n):
        U1 = rd.random()
        U2 = rd.random()
        R = math.sqrt(-2 * math.log(U1))
        t = 2 * math.pi * U2
        x = R * math.cos(t)
        y = R * math.sin(t)
        lista.append(x)
        lista.append(y)
    return lista

#Función para generar lognormales con parámetros de la normal subyacente
def lognormal_directa(mu, sigma2, n):
    #Generar Y ~ N(mu, sigma2)
    normales = box_muller(n)
    Y = [mu + math.sqrt(sigma2) * z for z in normales]
    #Transformar a lognormal X = e^Y
    X = [math.exp(y) for y in Y]
    return X

#Función para generar lognormales con media y varianza deseadas
def lognormal_desde_momentos(mu_prim, sigma2_prim, n):
    #Calcular parámetros de la normal subyacente
    mu = math.log(mu_prim**2 / math.sqrt(mu_prim**2 + sigma2_prim))
    sigma2 = math.log(1 + sigma2_prim / mu_prim**2)

    #Generar las lognormales
    X = lognormal_directa(mu, sigma2, n)
    return X, mu, sigma2

#Parámetros deseados para la lognormal
mu_prim = 5
sigma2_prim = 2

#Muestras
n_muestras = 100000
lognormales, mu, sigma2 = lognormal_desde_momentos(mu_prim, sigma2_prim, n_muestras//2)

#Gráfica
plt.hist(lognormales, bins=100, density=True, color="orange", alpha=0.5, edgecolor="black")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.title("Distribución Lognormal")
plt.show()

#Estadísticas empíricas
media_emp = np.mean(lognormales)
var_emp = np.var(lognormales)
desv_emp = math.sqrt(var_emp)

#Estadísticas teóricas
media_teo = mu_prim
var_teo = sigma2_prim
desv_teo = math.sqrt(var_teo)

#Resultados
print("\n--- Parámetros de la normal subyacente ---")
print(f"μ calculado: {mu:.4f}")
print(f"σ² calculado: {sigma2:.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media obtenida: {media_emp:.4f}")
print(f"Varianza obtenida: {var_emp:.4f}")
print(f"Desviación estándar obtenida: {desv_emp:.4f}")

print("\n--- Resultados teóricos ---")
print(f"Media deseada: {media_teo:.4f}")
print(f"Varianza deseada: {var_teo:.4f}")
print(f"Desviación estándar deseada: {desv_teo:.4f}")
```

**Pseudocodigo**
1. Generar $ Y_1 \sim \text{Gamma}(\alpha, 1) \text{ y } Y_2 \sim \text{Gamma}(\beta, 1) \text{ independiente de} Y_1$
2. Devolver  $X = \frac{Y_1}{Y_1 + Y_2}$

```python
#BETA
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta as beta_dist

def generar_beta(alpha, beta, n):
    """
    Genera muestras de una distribución Beta(alpha, beta) usando el método Gamma.

    Parámetros:
        alpha (float): Primer parámetro de forma (alpha > 0).
        beta (float): Segundo parámetro de forma (beta > 0).
        n (int): Número de muestras a generar.

    Devuelve:
        np.ndarray: Array de muestras Beta(alpha, beta).
    """
    if alpha <= 0 or beta <= 0:
        raise ValueError("alpha y beta deben ser positivos.")

    Y1 = np.random.gamma(alpha, 1, n)
    Y2 = np.random.gamma(beta, 1, n)
    return Y1 / (Y1 + Y2)

#Parámetros
alpha, beta, n = 0.5, 0.5, 1000
muestras = generar_beta(alpha, beta, n)

#Visualización
plt.hist(muestras, bins=100, density=True, color="purple", alpha=0.5, edgecolor="black")
x = np.linspace(0, 1, 1000)
plt.title('Distribución Beta({}, {})'.format(alpha, beta))
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.show()

#Resultados
print("\n--- Resultados empíricos ---")
print("Media estimada:", np.mean(muestras))
print("Varianza estimada:", np.var(muestras))
print("Desviación estándar estimada:", np.std(muestras))
```

```python
#TIPO V
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random
from math import gamma, log

def gamma_inversa(alpha, n):
    from scipy.stats import gamma as gamma_dist
    return gamma_dist.rvs(a=alpha, scale=1.0, size=n)

def pearson_type_v(alpha, beta, n):
    Y = gamma_inversa(alpha, n)
    X = beta / Y
    return X

#Parámetros
alpha = 3.0
beta = 2.0
n = 10000

#Simular datos
A = pearson_type_v(alpha, beta, n)

# Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='plum', edgecolor="black")
plt.title(f"Distribución Pearson Type V (Inversa Gamma)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()

#Estadísticas teóricas (válidas si α > 1 o α > 2)
media_teorica = beta / (alpha - 1) if alpha > 1 else np.nan
varianza_teorica = (beta ** 2) / ((alpha - 1) ** 2 * (alpha - 2)) if alpha > 2 else np.nan
desviacion_teorica = np.sqrt(varianza_teorica) if alpha > 2 else np.nan


print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}" if not np.isnan(media_teorica) else "Media teórica: No existe")
print(f"Varianza teórica: {varianza_teorica:.4f}" if not np.isnan(varianza_teorica) else "Varianza teórica: No existe")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}" if not np.isnan(desviacion_teorica) else "Desviación estándar teórica: No existe")

print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

```python
#LOG-LOGISTICA
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random


def loglogistica(alpha, beta, n):
    x = []
    for _ in range(n):
        u = random()
        valor = alpha * (u / (1 - u)) ** (1 / beta)  #Transformada inversa
        x.append(valor)
    return x

#Parámetros
alpha = 2.0
beta = 3.0
n = 10000


A = loglogistica(alpha, beta, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='green', edgecolor="black")
plt.title(f"Distribución Log-logística LL(α={alpha}, β={beta})")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()


if beta > 1:
    media_teorica = (np.pi * alpha / beta) / np.sin(np.pi / beta)
else:
    media_teorica = np.nan  # No existe

if beta > 2:
    varianza_teorica = (alpha ** 2) * (2 * np.pi / beta) / np.sin(2 * np.pi / beta) - media_teorica ** 2
    desviacion_teorica = np.sqrt(varianza_teorica)
else:
    varianza_teorica = np.nan
    desviacion_teorica = np.nan


print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}" if not np.isnan(media_teorica) else "Media teórica: No existe")
print(f"Varianza teórica: {varianza_teorica:.4f}" if not np.isnan(varianza_teorica) else "Varianza teórica: No existe")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}" if not np.isnan(desviacion_teorica) else "Desviación estándar teórica: No existe")

print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

**Pseudocodigo**

```python
#JOHNSON
#Librerías
import numpy as np
import matplotlib.pyplot as plt

def johnson_su_simulacion(gamma, delta, xi, lambd, n):
    Z = np.random.normal(0, 1, n)  #Z ~ N(0, 1)
    X = xi + lambd * np.sinh((Z - gamma) / delta)
    return X

#Parámetros
gamma = 0.0
delta = 1.0
xi = 0.0
lambd = 1.0
n = 10000

#Simulacion
A = johnson_su_simulacion(gamma, delta, xi, lambd, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='mediumseagreen', edgecolor='black')
plt.title("Distribución Johnson SU (simulada desde normal estándar)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()


print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

```python
#GEOMETRICA
#Libreias
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math

def geometrica(p, n):
    x = []
    for _ in range(n):
        U = random()
        val = math.ceil(math.log(1 - U) / math.log(1 - p))
        x.append(val)
    return x

#Simulaciones
A = geometrica(0.3, 10000)

#Histograma
plt.hist(A, bins=np.arange(0.5, max(A)+1.5, 1), density=True, edgecolor='black', color='orange', alpha=0.5)
plt.title("Distribución Geométrica (p=0.3)")
plt.show()

media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)

#Resultados
print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

\begin{equation}
\sigma =  \frac{\sqrt {s(1-p)}}{p}

**Pseudocodigo**

1. Generar $ U \sim U(0,1)$ \\
2. Buscar el menor  $x $ tal que\\
   $\quad \sum_{k=0}^{x} \binom{k+r-1}{k} p^r (1-p)^k \geq U$ \\
3. Retornar  $ x $



```python
#BINOMIAL NEGATIVA
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math

def binomial_negativa(r, p, n):
    x = []
    for _ in range(n):
        total = 0
        for _ in range(r):
            U = random()
            g = math.ceil(math.log(1 - U) / math.log(1 - p))
            total += g - 1
        x.append(total)
    return x

#Simulación
A = binomial_negativa(3, 0.3, 10000)

#Histograma
plt.hist(A, bins=np.arange(min(A) - 0.5, max(A) + 1.5, 1), density=True, edgecolor='black', color='plum')
plt.title("Distribución Binomial Negativa (r=3, p=0.3)")
plt.show()

media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)

#Resutados
print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

**Poisson** $(\lambda)$

Función de densidad
\begin{equation}
f(x)=
\begin{cases}
\frac{e^{- \lambda } λ^x}{x !} &  \text{si } x \leq 0\\
0 & otro \;\; caso
\end{cases}
\end{equation}

Funcion Acumulativa
\begin{equation}
F(x)=
\begin{cases}
0  & \text{si } x < 0\\
e^{- \lambda} \sum \frac{\lambda ^i}{i!} & \text{si } x \geq 0
\end{cases}
\end{equation}

Varianza

\begin{equation}
\sigma =  \sqrt{\lambda}

**Pseudocodigo**


1. Generar $ U \sim U(0,1),\ F = e^{-\lambda},\ x = 0$ \\
2. Mientras $ F < U:$ \\
   $\quad x = x + 1 $\\
   $\quad F = F + \frac{e^{-\lambda} \lambda^x}{x!}$ \\
3. Retornar $ x$



```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math

def poisson(lamb, n):
    x = []
    for _ in range(n):
        U = random()
        k = 0
        p = math.exp(-lamb)
        F = p
        while U > F:
            k += 1
            p *= lamb / k
            F += p
        x.append(k)
    return x

#Simulaciones
A = poisson(4, 10000)

#Histograma
plt.hist(A, bins=np.arange(min(A)-0.5, max(A)+1.5, 1), density=True, edgecolor='black', alpha=0.5)
plt.title("Distribución Poisson (λ = 4)")
plt.show()

media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)

#Resultados
print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```
