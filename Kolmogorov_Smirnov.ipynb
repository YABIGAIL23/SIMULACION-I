```python
import numpy as np
from scipy.stats import kstest, ksone
import pandas as pd
from tabulate import tabulate


datos = sorted([
    0.048, 0.591, 0.478, 0.166, 0.932,
    0.926, 0.007, 0.551, 0.362, 0.533,
    0.368, 0.437, 0.178, 0.411, 0.545,
    0.787, 0.594, 0.086, 0.951, 0.772,
    0.298, 0.199, 0.157, 0.401, 0.631
]) #Datos ordenados

n = len(datos)
alpha = 0.05
indices = np.arange(1, n + 1)
f_n = indices / n  #i/n
f_n_minus = (indices - 1) / n  #(i-1)/n
D_plus = f_n - datos  #i/n - X_i
D_minus = datos - f_n_minus   #X_i - (i-1)/n

#valores críticos
D_plus_max = np.max(D_plus)
D_minus_max = np.max(D_minus)
D = max(D_plus_max, D_minus_max)
D_critico = ksone.ppf(1 - alpha / 2, n)

#Crear DataFrame con valores redondeados
tabla = pd.DataFrame({
    'i': indices,
    'X_i': np.round(datos, 5),
    'i/n': np.round(f_n, 5),
    '(i-1)/n': np.round(f_n_minus, 5),
    'i/n - X_i': np.round(D_plus, 5),
    'X_i - (i-1)/n': np.round(D_minus, 5)
})

#Tabla
print("\nTabla de Cálculo KS:")
print(tabulate(tabla, headers='keys', tablefmt='fancy_grid', showindex=False))

#Prueba de Kolmogorov-Smirnov
statistic, p_value = kstest(datos, 'uniform')

#Imprimir resultados
print(f"\nD crítico: {D_critico:.5f}")
print(f"Estadístico KS: {statistic:.5f}")
print(f"Valor p: {p_value:.5f}")
print(f"\nD+ = {D_plus_max:.5f}")
print(f"D- = {D_minus_max:.5f}")
print(f"D = {D:.5f}")

#Conclusión de la prueba
if D > D_critico:
    print("\033[91mRechazamos la hipótesis nula: Los datos no siguen una distribución uniforme.\033[0m")
else:
    print("\033[92mNo se puede rechazar la hipótesis nula: Los datos podrían seguir una distribución uniforme.\033[0m")
```
