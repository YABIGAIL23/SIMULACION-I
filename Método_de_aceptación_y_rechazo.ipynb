### Método de Aceptación y Rechazo
Es un método generador de números aleatorios que permite simular una variable aleatoria con una distribución deseada, cuando no es posible muestrearla directamente.


```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parámetros
a = 1
b = 3
c = 5
```

Función de densidad de la distribución triángular

image.png

```python
def f(x):
    if a <= x <= b:
        return (2*(x - a)) / ((c - a)*(b - a))
    elif b < x <= c:
        return (2*(c - x)) / ((c - a)*(c - b))
    else:
        return 0
```

```python
#Valor máximo de la función f(x)
M = 2 / (c - a)
```

```python
#Método
def aceptacion(n):
    lista = []
    while len(lista) < n:
        R1 = rd.random()
        R2 = rd.random()
        x = a + (c - a) * R1
        if R2 <= f(x) / M:
            lista.append(x)
    return lista
```

```python
muestras = aceptacion(50000)
```

```python
#Media y desviación estándar
media = np.mean(muestras)
desviacion = np.std(muestras)
print("Media: ",media)
print("Desviación estándar: ",desviacion)
```

```python
#Gráfica
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Histograma muestras')
plt.show()
```
