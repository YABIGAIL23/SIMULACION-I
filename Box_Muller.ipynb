###Método Box-Muller
El método de Box-Muller, propuesto por George E. P. Box y Mervin E. Muller en 1958, es una técnica matemática para generar números aleatorios con una distribución normal estándar (media 0 y desviación estándar 1), a partir de dos números aleatorios con distribución uniforme en (0, 1).

```python
#Librerias
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math
```

    Genera 2n números aleatorios con distribución normal estándar
    utilizando el método de Box-Muller.
    
    Parámetro:
    n -- número de pares de números normales a generar
    
    Retorna:
    lista de 2n números con distribución normal estándar

image.png

```python
#Método de Box-Muller
def box_muller(n):
    lista = []
    for i in range(n):
        U1 = rd.random()
        U2 = rd.random()
        R = math.sqrt(-2 * math.log(U1))
        t = 2 * math.pi * U2
        x = R * math.cos(t)
        y = R * math.sin(t)
        lista.append(x)
        lista.append(y)
    return lista
```

```python
#Muestras
m = box_muller(100000)  #Genera 200000 números
```

```python
#Grafica
plt.hist(m, bins=100, density=True)
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

```python
#Estadísticas
media = np.mean(m)
desviacion = np.std(m)
print("Media obtenida: ", media)
print("Desviación estándar obtenida: ", desviacion)

# Comparación con valores teóricos
print("Media teórica: 0")
print("Desviación estándar teórica: 1")
```
