###M√âTODO DE TRANSFORMADA INVERSA
Es un m√©todo que permite generar n√∫meros aleatorios que siguen una distribuci√≥n espec√≠fica (como la triangular), partiendo de una variable uniforme

```python
#Librer√≠as
import numpy as np
import matplotlib.pyplot as plt
from random import random
```


$ x = \begin{cases}
a + \sqrt{R \cdot (b - a)(c - a)} & \text{si } R < \frac{b-a}{c-a} \\
c - \sqrt{(1 - R) \cdot (c - b)(c - a)} & \text{si } R \geq \frac{b-a}{c-a}
\end{cases} $

```python
#Funci√≥n de la transformada inversa para la distribucion triangular
def triangular(n):
  a,b,c=1,3,5 #Parametros
  x = [] #Lista vacia
  for i in range(n):
    R = random() #Numeros Aleatorios
    if R<(b-a)/(c-a):
      x.append(a + np.sqrt(R * (b - a) * (c - a)))
    elif R>=(b-a)/(c-a):
      x.append(c - np.sqrt((1 - R) * (c - b) * (c - a)))
  return x
```

```python
#Muestras
A = triangular(10000)
```

```python
#Histograma
plt.hist(A, bins=100, density=True)
plt.title("Distribuci√≥n Triangular")
```

```python
#Media y varianza
print("Media:", np.mean(A))
print("Varianza:", np.var(A))
```
