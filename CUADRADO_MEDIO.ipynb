```python
from tabulate import tabulate

def cuadrado_medio(semilla, n):
    resultados = []
    valores_vistos = {}  #Para detectar ciclos
    valor = semilla
    ciclo_inicio = None

    for i in range(n):
        cuadrado = valor ** 2
        cuadrado_str = str(cuadrado).zfill(8)  #Siempre 8 dígitos
        mitad = len(cuadrado_str) // 2

        # Obtener los 4 dígitos
        nuevo_numero_str = cuadrado_str[mitad - 2 : mitad + 2]
        valor = int(nuevo_numero_str)

        #Normalización
        numero_normalizado = valor / 10000

        #Detectar ciclo
        if valor in valores_vistos:
            ciclo_inicio = valores_vistos[valor]
            break

        valores_vistos[valor] = i
        resultados.append((valor, numero_normalizado))

    return resultados, ciclo_inicio

#Validación de semilla
while True:
    semilla = input("Ingresa la semilla (número entero de 4 dígitos): ")
    if semilla.isdigit() and len(semilla) == 4:
        semilla = int(semilla)
        break
    else:
        print("ERROR: Ingresa solo 4 dígitos.")

#Validación de cantidad
while True:
    try:
        cantidad = int(input("¿Cuántos números deseas generar?: "))
        if cantidad > 0:
            break
        else:
            print("ERROR: Ingresa un número mayor que 0.")
    except ValueError:
        print("ERROR: Debes ingresar un número entero.")

#Generar números
numeros, ciclo_inicio = cuadrado_medio(semilla, cantidad)

#Preparar datos para la tabla
tabla = []
for i, (num, norm) in enumerate(numeros):
    estado = "Inicio del Ciclo" if ciclo_inicio is not None and i == ciclo_inicio else ""
    tabla.append([i + 1, num, norm, estado])

#Mostrar tabla con encabezados
print("\nResultados generados con el método del cuadrado medio:")
print(tabulate(tabla, headers=["Iteración", "Número generado", "Número Normalizado", "Estado"], tablefmt="fancy_grid"))

#Informar si se detectó ciclo
if ciclo_inicio is not None:
    print(f"\nCiclo detectado desde la iteración {ciclo_inicio + 1} con longitud {len(numeros) - ciclo_inicio}.")
else:
    print("\n No se detectaron ciclos.")
```
