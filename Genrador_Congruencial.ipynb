```python

#Generador congruencial

import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt

a = 81
c = 89
m = 100
x = 5

def generar_numeros(x, a, c, m, n):
    num = []
    tabla = []
    for i in range(n):
        x = (a * x + c) % m
        num.append(x)
        tabla.append([i+1 , x, x / m])

    print(tabulate(tabla, headers=["Iteración", "x", "x/m"], tablefmt="grid"))
    return num

numeros_generados = generar_numeros(x, a, c, m, 100)

lista = [num / m for num in numeros_generados]

plt.figure(figsize=(10, 5))
plt.plot(lista, marker='o', linestyle='-', color='lightblue')
plt.xlabel('Iteración')
plt.ylabel('x/m')
plt.title('Números generados normalizados')
plt.grid()
plt.show()

plt.hist(lista, color='pink', bins=10)  # Reduciendo bins para mejor visualización
plt.xlabel('Valor de x/m')
plt.ylabel('Frecuencia')
plt.title('Histograma de números generados')
plt.grid()
plt.show()
```
